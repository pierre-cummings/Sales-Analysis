```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("superstore.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
```

```python
monthly_sales.plot()
plt.title('Monthly Sales Trend')
plt.show()
```

```python
# Sales are generally increased during Q4 due to holiday season.
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
category_sales.plot(kind='bar')
plt.title('Sales by Category')
plt.show()
```

```python
#Technology generates the highest revenue
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
region_sales.plot(kind='bar')
plt.title('Sales by Region')
plt.show()
```

```python
#Highest demand is in the Western region
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar')
plt.title('Top Product by Sales')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Canon Adv Copier by far Top Product in Demand
```
